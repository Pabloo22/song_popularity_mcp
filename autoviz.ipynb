En este notebook vamos a visualizar los datos de manera automática usando
la librería autoviz. Esta librería nos proporciona, además, un informe
con sugerencias de transformaciones.

```python
from autoviz import AutoViz_Class
```

```python
AV = AutoViz_Class()
dft = AV.AutoViz('data/train.csv', sep=",", depVar='song_popularity', verbose=2, 
                 lowess=False, chart_format='html', max_rows_analyzed=150000, max_cols_analyzed=30)
```
