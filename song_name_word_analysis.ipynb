```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd

from utils import load_data
```

```python
train, X_test = load_data(no_duplicates=True, split=False)

song_names = train['song_name']
join_words = ' '.join(song_names)
# Creamos la nube de palabras
wordcloud = WordCloud(width=1600, height=800, max_words=100, background_color='white').generate(join_words)

# Mostramos la nube de palabras
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
song_names.str.lower().value_counts()
```

```python
word_list = join_words.lower().split()
counter = Counter(word_list)
word_freq = pd.Series(counter)
word_freq.sort_values(ascending=False)
```

```python
word_freq.sort_values(ascending=False) / len(train)
```

```python
(word_freq.sort_values(ascending=False)[:20]).plot(kind='bar')
plt.title("Top 20 palabras más frecuentes")
plt.ylabel("número total de apariciones")
plt.show()
```
